```python
#TASK - CALCULATE ANY FIELDS IN THE NEW LAYERS THAT ARE JUST A ONE TO ONE EQUAL WITH OLD FIELDS ADDED
import csv
import arcpy

# Function to calculate fields for a given feature class
def calculate_fields(feature_class):
    try:
        # List of fields to calculate
        fields_to_calculate = [('UNITCODE', 'ALPHA'), ('BNDY_TYPE', 'BNDY_TYPE1'), ('MAPSOURCE', 'SOURCEDOC1'),
                               ('NOTES', 'DESCRIPT1'), ('NOTES1', 'DESCRIPT2'), ('MONSTAMPID', 'MON_STAMP')]

        # Iterate over fields to calculate
        for output_field, input_field in fields_to_calculate:
            arcpy.CalculateField_management(in_table=feature_class,
                                            field=output_field,
                                            expression=f"!{input_field}!",
                                            expression_type='PYTHON3')

        print(f"Calculation of fields for {feature_class} completed successfully.")

    except arcpy.ExecuteError:
        print(arcpy.GetMessages(2))
    except Exception as ex:
        print(ex)

try:

    # Set the workspace
    arcpy.env.workspace = target_feature_dataset

    # Create a list to store layer names from "Layer_new" column in the CSV
    layer_new_list = []

    # Read the layer mapping from the CSV file and populate the list with layer names
    with open(layer_map_csv, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            layer_new_list.append(row['Layer_new'])

    # Iterate over each layer name in the list
    for layer_new in layer_new_list:
        # Construct the full path of the feature class in the target feature dataset
        feature_class_path = f'{target_feature_dataset}\\{layer_new}'

        # Check if the feature class exists
        if arcpy.Exists(feature_class_path):
            # Calculate fields for the feature class
            calculate_fields(feature_class_path)

except Exception as ex:
    print(f"An error occurred: {ex}")
```

```python
# TASK - CALCULATE DIRECTION FIELD FROM JOINED DIRECTION FIELD 
import arcpy
import re
import csv

# Function to calculate azimuth
def calculate_azimuth(value):
    print("Input value:", value)  # Debug print
    # Convert the value to a string
    value_str = str(value)
    print("Value as string:", value_str)  # Debug print

    pattern = re.compile(r"[NS](\d+)-(\d+)-(\d+)[EW]")
    match = pattern.match(value_str)
    if match:
        value1 = int(match.group(1))
        value2 = int(match.group(2))
        value3 = int(match.group(3))
        start_direction = value_str[0]
        end_direction = value_str[-1]
        if start_direction == 'N' and end_direction == 'E':
            bearing = value1
        elif start_direction == 'S' and end_direction == 'E':
            bearing = 180 - value1
        elif start_direction == 'S' and end_direction == 'W':
            bearing = 180 + value1
        elif start_direction == 'N' and end_direction == 'W':
            bearing = 360 - value1
        else:
            raise ValueError("Invalid direction combination")
        minute = value2 / 60
        second = value3 / 3600
        azimuth = bearing + minute + second
        return azimuth  # Return azimuth as double
    else:
        print("Pattern not found in direction:", value_str)
        return None

try:
    # Set the workspace
    arcpy.env.workspace = target_feature_dataset

    # Read the layer mapping from the CSV file and populate the list with layer names
    layer_new_list = []
    with open(layer_map_csv, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            if 'COGO_UseLocations_ln' in row['Layer_new'] or 'COGO_SurveyBndy_ln' in row['Layer_new']:
                layer_new_list.append(row['Layer_new'])

    # Iterate over each layer name in the list
    for layer_name in layer_new_list:
        # Construct the full path of the feature class in the target feature dataset
        feature_class_path = f'{target_feature_dataset}\\{layer_name}'

        # Use an update cursor to calculate azimuth and update the "Direction1" field
        with arcpy.da.UpdateCursor(feature_class_path, ['Direction1', 'Direction2']) as cursor:
            for row in cursor:
                azimuth = calculate_azimuth(row[1])  # Pass direction value from Direction2 field
                if azimuth is not None:
                    row[0] = azimuth  # Assign azimuth value as double to Direction1 field
                    cursor.updateRow(row)
        print(f"Azimuth calculated and updated for '{layer_name}'.")

except arcpy.ExecuteError as e:
    print(f"Error: {e}")
except Exception as ex:
    print(f"Exception: {ex}")
```
